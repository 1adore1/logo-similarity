```python
import torch
import torchvision
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as transforms
    
class LogoEncoder(nn.Module):
    def __init__(self, embedding_dim=512):
        super(LogoEncoder, self).__init__()
        self.model = torchvision.models.resnet50(weights=True)
        self.model.fc = nn.Linear(self.model.fc.in_features, embedding_dim)
        self.normalize = nn.functional.normalize

    def forward(self, x):
        x = self.model(x)
        return self.normalize(x, p=2, dim=1)
```

```python
device = torch.device('cpu')

model = LogoEncoder(embedding_dim=1024).to(device) 
model.load_state_dict(torch.load('models/best_model1024.pth', map_location=device))

model.eval()
test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

def load_image(image_path):
    img = Image.open(image_path).convert("RGB")
    img = test_transform(img)
    return img.to(device)

def get_similarity(ref_path, test_path, model):
    references = torch.stack([load_image(os.path.join(ref_path, img)) for img in os.listdir(ref_path)], dim=0)

    with torch.no_grad():
        embeddings = model(references)
        test_img = load_image(test_path).unsqueeze(0)
        test_emb = model(test_img)
    
    mx = max(F.cosine_similarity(emb, test_emb) for emb in embeddings)
    return mx

def show_imgs(ref_path, test_path):
    references = torch.stack([load_image(os.path.join(ref_path, img)) for img in os.listdir(ref_path)], dim=0)
    test_img = load_image(test_path).permute(1, 2, 0).numpy()

    num_images = min(5, len(references))
    fig, axes = plt.subplots(1, num_images + 1, figsize=(10, 10))

    for i, img in enumerate(references[:num_images]):
        ref_img = img.permute(1, 2, 0).numpy()

        axes[i].imshow(ref_img)
        axes[i].set_title(f'reference {i+1}')
        axes[i].axis('off')

    axes[i+1].imshow(test_img)
    axes[i+1].set_title(f'test image')
    axes[i+1].axis('off')

    plt.show()

for ref_folder in os.listdir(os.path.join('test_cases', 'ref_imgs')):
    ref_path = os.path.join(os.path.join('test_cases', 'ref_imgs'), ref_folder)

    for test_file in os.listdir(os.path.join('test_cases', 'test_imgs')):
        test_path = os.path.join(os.path.join('test_cases', 'test_imgs'), test_file)
        if os.path.isfile(test_path):
            show_imgs(ref_path, test_path)
            sim = get_similarity(ref_path, test_path, model)
            if sim > 0.7:
                print(f'Objects match, similarity: {sim.item():.4f}')
            else:
                print(f"Objects don't match, similarity: {sim.item():.4f}")

```
